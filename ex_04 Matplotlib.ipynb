```python
# plot 함수 - 일직선으로 되어있는 Lineplot그리기
y=[2,4,6]

plt.plot(y)  # 그래프를 그리기
plt.show()  # 그래프를 출력
```

```python
# x축 설정하여 그래프 그리기
x=[1,2,3]
y=[2,4,6]

plt.plot(x, y)
plt.show()
```

```python
# line style -> ls
# 그려지는 라인의 스타일 지정
plt.plot(x,y, ls='--')
plt.show()
```

```python
x=[2,4,6,8]
y=[10,13,16,20]
y2=[10,20,16,13]
plt.plot(x,y)
plt.plot(x,y2, ls =':')
plt.show()
```

```python
plt.plot(x,y)
plt.show()
plt.plot(x,y2)
plt.show()
```

```python
# marker
plt.plot(x,y, marker='o')
plt.show()
```

```python
# line width ->lw
# 라인 두께 지정
plt.plot(x,y, lw='5')
plt.show()
```

```python
# 선 색지정
# 마커 색지정
# ->  markerfacecolor -> mfc
plt.plot(x,y, color='r', marker='o', mfc='b')
plt.show()
```

```python
import numpy as np
x= np.arange(7)
y=[1,4,5,8,9,5,3]
plt.plot(x,y, lw=5, marker='o',ls='--', ms=15,  mfc='r', mec='g', mew=5)
plt.show()
```

```python
plt.plot(x,y, lw=5, marker='o',ls='--', ms=15,  mfc='r', mec='g', mew=5)
plt.xlim(-2,8) # x범위 지정(최소, 최대)
plt.ylim(-1,11) # y범위 지정(최소, 최대)
plt.show()
```

```python
x= np.arange(1,51) #1~50 array
y= np.random.randint(1,50, size=50) # 1~50사이 50개의 랜덤수
print(x.shape)
print(y.shape)

plt.plot(x,y)
plt.ylim(30,50) # y범위 지정(최소, 최대)
plt.xlim(20,40) # x범위 지정(최소, 최대)
plt.grid() #눈금선 설정
plt.show()
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
x= np.arange(7)
y=[1,4,5,8,9,5,3]
plt.plot(x,y, lw=5, marker='o',ls='--', ms=15,  mfc='r', mec='g', mew=5)
plt.xticks([0,3,6])
plt.yticks([1,4,7])
plt.grid()
plt.show()
```

```python
x= np.arange(7)
y=[1,4,5,8,9,5,3]
plt.plot(x,y, lw=5, marker='o',ls='--', ms=15,  mfc='r', mec='g', mew=5)
plt.xticks([0,3,6])
plt.yticks([1,4,7])
plt.xlabel('x') # x축의 이름
plt.ylabel('y', rotation=0) # y축의 이름 #rotation= 라벨 위치 돌리기
plt.grid()
plt.show()
```

```python
# 한 영역 안에 여러 그래프 그리기
x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9,12]

plt.plot(x, color='r', label='x') # 그래프 이름
plt.plot(y, color='k', label='y')
plt.plot(z, color='m', label='z')
plt.legend(loc='best') # 범례 띄우기
plt.show()

#   best
#   upper right
#   upper left
#   lower left
#   lower right
#   right
#   center left
#   center right
#   lower center
#   upper center
#   center
```

```python
plt.plot(x, color='r', label='엑스') # 그래프 이름
plt.plot(y, color='k', label='와이')
plt.plot(z, color='m', label='제트')
plt.xlabel('x축', color='r')
plt.ylabel('y축', rotation=0, color='r')
plt.title('x,y,z 그래프', color='r')
plt.legend(loc='best') # 범례 띄우기
plt.show()
```

#### 차대차 사건중 죽거나 다친 사람이 많은 발생지 시도를 알아보고 시각화

```python
# 1. 차태차 사건 중
# 2. (!) 죽거나 다친 사람(사상자)이 많은 (2) 발생자 시도
# 3. 시각화
```

```python
data=pd.read_csv('Traffic_Accident_2017.csv',encoding="euc-kr")
data
```

```python
data.head(2)
```

```python
# 1. 차태차 사건
data['사고유형_대분류']=='차대차' # 불리언 (T/F)
acc=data[data['사고유형_대분류']=='차대차']
acc
```

```python
# 2. (1)죽거나 다친 사람(사상자)이 많은  (2) 발생지 시도
acc['발생지시도'].unique()
```

```python
acc['사상자수'].value_counts()
```

```python
# 그룹화 - groupby('그룹화 시킬 컬럼').sun()
acc_g=acc[['사상자수', '발생지시도']].groupby('발생지시도').sum()
acc_g
```

```python
x=acc_g.index
y=acc_g['사상자수']
```

```python
# 시각화
plt.figure(figsize=(10,5)) #(가로, 세로)
plt.xlabel('지역', color='y')
plt.ylabel('사상자수', rotation=0, color='y')


plt.bar(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9,12]

plt.figure(figsize=(10,5))

plt.plot(x, color='r', label='x') # 그래프 이름
plt.plot(y, color='k', label='y')
plt.plot(z, color='m', label='z')
plt.legend(loc='best') # 범례 띄우기
plt.show()
```
